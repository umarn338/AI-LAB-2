# Task 1: NumPy Array Creation


```python
import numpy as np


arr1 = np.arange(10)
print("1D Array:", arr1)


arr2 = np.arange(1, 10).reshape(3, 3)
print("\n2D Array (3x3):\n", arr2)


arr3 = np.random.randint(1, 101, size=(2, 3, 4))
print("\n3D Array (2x3x4):\n", arr3)
```

Task 2: Array Indexing and Slicing


```python

second_row = arr2[1, :]
print("\nSecond row of arr2:", second_row)


element = arr2[2, 1]
print("\nElement at 3rd row and 2nd column of arr2:", element)


first_layer = arr3[0, :, :]
print("\nElements from 0th layer of arr3:\n", first_layer)
```

Task 3: Array Reshaping


```python

reshaped_arr1 = arr1.reshape(2, 5)
print("\nReshaped 1D Array to 2x5:\n", reshaped_arr1)
```

Task 4: Mathematical Operations on Arrays

```python

arr1_plus_10 = arr1 + 10
print("\nArray after adding 10 to each element:\n", arr1_plus_10)


arr2_times_2 = arr2 * 2
print("\nArray after multiplying each element by 2:\n", arr2_times_2)
```

# Scatter Plot

```python
x_scatter = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y_scatter = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, color='black', s=100, alpha=0.6)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')
plt.show()
```

# Bar Plot

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]
colors = ['blue', 'green', 'red', 'purple', 'orange']

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

# Histogram

```python
data = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')
plt.show()
```

# Pie Chart

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]
explode = (0.1, 0, 0, 0)  # Explode the first slice (Python)

plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.title('Pie Chart of Programming Languages')
plt.show()
```
